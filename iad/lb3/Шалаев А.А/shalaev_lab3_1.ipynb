```python
from tensorflow.keras.applications.densenet import DenseNet201
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet import preprocess_input, decode_predictions
import numpy as np
```

```python
model = DenseNet201(weights = 'imagenet')
```

```python
# VGG16
img = image.load_img('/content/drive/My Drive/Colab/cello.png', target_size=(224, 224))

# Xception, InceptionV3
# img = image.load_img('/content/drive/My Drive/plane.jpg', target_size=(299, 299))

# MobileNet, DenseNet
#img = image.load_img('/content/drive/My Drive/plane.jpg', target_size=(224, 224))

# ResNet
#img = image.load_img('/content/drive/My Drive/plane.jpg', target_size=(224, 224))

# EfficientNetB0
# img = image.load_img('/content/drive/My Drive/plane.jpg', target_size=(224, 224))

# NasNetLarge
# img = image.load_img('/content/drive/My Drive/plane.jpg', target_size=(331, 331))
```

```python
img
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
preds = model.predict(x)
# preds.shape
```

```python
# Печатаем 3 наиболее вероятных предсказания
print('Результаты распознавания: ', decode_predictions(preds, top=4)[0])
```
