```python
ModelName='Facenet'
DetectorBackend='dlib'
```

```python
def detect_face(img):
    detected_face = DeepFace.detectFace(img, detector_backend = DetectorBackend)
    plt.imshow(cv2.cvtColor(detected_face, cv2.COLOR_BGR2RGB))
```

```python
def detect_face_orig(img):
    orig = cv2.imread(img)
    plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
```

```python
def verify_images(img1, img2):
    obj = DeepFace.verify(img1, img2, distance_metric='euclidean', model_name=ModelName, detector_backend = DetectorBackend)
    return obj
 
```
