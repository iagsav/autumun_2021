```python
import shutil
import os
```

```python
data_dir = '/content/PetImages/'
train_dir = 'train'
val_dir = 'val'
test_dir = 'test'
test_portion = 0.15
val_portion = 0.15
nb_images = 12500
```

```python
def create_dir(dir_name):
  if os.path.exists(dir_name):
    shutil.rmtree(dir_name)
  os.makedirs(dir_name)
  os.makedirs(os.path.join(dir_name, "cats"))
  os.makedirs(os.path.join(dir_name, "dogs"))
```

```python

create_dir(train_dir)
create_dir(test_dir)
create_dir(val_dir)
```

```python
def copy_images(st, ed, src, dst):
  for i in range(st, ed):
    shutil.copy2(os.path.join(src, "Cat/" + str(i) + ".jpg"), os.path.join(dst, "cats"))

    shutil.copy2(os.path.join(src, "Dog/" + str(i) + ".jpg"), os.path.join(dst, "dogs"))
```

```python
start_val = int(nb_images * (1 - val_portion - test_portion))
start_tst = int(nb_images * (1 - test_portion))
print(start_val)
print(start_tst)
```

```python
copy_images(0, start_val, data_dir, train_dir)
```

```python
copy_images(start_val, start_tst, data_dir, val_dir)
```

```python
copy_images(start_tst, nb_images, data_dir, test_dir)
```

```python
#%tensorflow_version 2.x
from tensorflow.keras.models import Sequential
from keras.layers.convolutional import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Dropout, Flatten, Activation, Dense
from keras.preprocessing.image import ImageDataGenerator
```

```python
img_width, img_height = 150, 150
input_shape = (img_width, img_height, 3)
epochs = 3
batch_size = 128
nb_train = 17500
nb_val   = 3750
nb_test  = 3750
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape = input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer= 'adam', metrics = ['accuracy'])
```

```python
datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_generator = datagen.flow_from_directory(train_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = batch_size,
                                              class_mode='binary')
```

```python
val_generator = datagen.flow_from_directory(val_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = batch_size,
                                              class_mode='binary')
```

```python
test_generator = datagen.flow_from_directory(test_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = batch_size,
                                              class_mode='binary')
```

```python
train_dir = 'HuH/train'
val_dir='HuH/val'
test_dir='HuH/test'
img_width,img_height=150,150
input_shape=(img_width,img_height,3)
epochs=10
bath_size=20
nb_train_samples=35
nb_val_samples=35
nb_test_samples=35
```

```python
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import ImageDataGenerator
```

```python

datagen=ImageDataGenerator()
```

```python
 val_generator = datagen.flow_from_directory(val_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = bath_size,
                                              class_mode='binary')
```

```python
train_generator = datagen.flow_from_directory(train_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = bath_size,
                                              class_mode='binary')
```

```python

test_generator = datagen.flow_from_directory(test_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = bath_size,
                                              class_mode='binary')
```

```python
from tensorflow.keras.applications.mobilenet import MobileNet
```

```python
mobilenet_net=MobileNet(weights='imagenet',include_top=False,input_shape=(150,150,3))
mobilenet_net.trainable=False
```

```python
mobilenet_net.summary()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Flatten, Activation, Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
model.add(mobilenet_net)
model.add(Flatten())
model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
optimizer=Adam(lr=1e-5),
metrics=['accuracy'])
```

```python
model.fit_generator(train_generator,
                    steps_per_epoch=nb_train_samples,
                    epochs=10,
                    validation_data=val_generator,
                    validation_steps=nb_val_samples)
```

```python
 Scores=model.evaluate_generator(test_generator,nb_test_samples)
 print("",Scores[1]*100)
```
