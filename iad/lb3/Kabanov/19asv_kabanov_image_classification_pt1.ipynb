```python
%tensorflow_version 2.x
from tensorflow.keras.applications.resnet import ResNet101
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet import preprocess_input, decode_predictions
import numpy as np
```

```python
# Загружаем сеть
model = ResNet101(weights = 'imagenet')
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python

# ResNet
img = image.load_img('/content/drive/MyDrive/UNIBS/ColabN/Beagle_745.jpg', target_size=(224, 224))
```

```python
img
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
preds = model.predict(x)
# preds.shape
```

```python
# Печатаем 3 наиболее вероятных предсказания
print('Результаты распознавания: ', decode_predictions(preds, top=3)[0])
```
