Удаление столбца ID

```python
data = data.drop(columns=['Id'])#Удаление колонки id
data.head()
```

```python
data_object = data.select_dtypes(include=[object])#Выбор столбцов по типу данных:объекты
data_object
```

```python
import numpy as np
data_int = data.select_dtypes(include=[np.number])#Выбор столбцов по типу данных:числовые
data_int
```

```python
data_object = data_object.dropna(axis = 'columns')#Удаление столбцов, в котором остутсвует хотя бы один элемент
data_object
```

```python
from sklearn import preprocessing #модуль для предварительной обработки данных
le = preprocessing.LabelEncoder()
encoded = data_object.apply(le.fit_transform)#Замена текстовых данных новыми закодированными
encoded.head()
```

```python
encoded.reset_index(drop=True, inplace=True)#reset_index () устанавливает список целых чисел от 0 до длины данных в качестве индекса
data_int.reset_index(drop=True, inplace=True)
```

```python
data_1 = pd.concat([encoded, data_int], axis = 1)#Объединение двух фреймов данных 
data_1.head()
```

```python
# Среднее значение
mean = data_1.mean(axis=0)
# Стандартное отклонение
std = data_1.std(axis=0)
data_1 -= mean
data_1 /= std
data_1.head()
```

```python
data_1.describe()# выводит стандартное отклонение набора данных
```

```python
y = data_1.iloc[:, 63]#Выбор определенного столбца
y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split #Разделение данных для обучения и тестирования
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.NaN, strategy='mean')#Замена пропущенных значений средним значанием
data_2 = pd.DataFrame(imp.fit_transform(data_1))
data_2.columns=data_1.columns
data_2.index=data_1.index
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_2, y, test_size=0.30, random_state=10)#test_size используется для разбиения выборки в нашем случае 30% не тесты
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

model = Sequential()
model.add(Dense(256, activation="relu", input_shape=(x_train.shape[1],)))
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam", loss="mse", metrics=["mae"])#model.compile создает модель и принимает параметр 'metrics', чтобы определить, какие метрики используются для оценки во время обучения и тестирования
model.summary()
history = model.fit(x_train, y_train, epochs=50, batch_size=20, validation_split=0.2)#Мы передаем валидационные данные для мониторинга потерь и метрик на этих данных в конце каждой эпохи
```

```python
pred = model.predict(x_test)
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
pred = pred.flatten()
np.corrcoef(pred, y_test)
```
