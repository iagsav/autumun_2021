```
import pandas as pd
data = pd.read_csv("/content/train.csv")
```

```
data
```

```
data = data.drop(columns=['id'])
data = data.drop(columns=['timestamp'])
data
```

```
# Удаление строк с NAN-значениями
data = data.dropna()
data
```

```
cols = data.columns.tolist()
cols
```

```
# Перемещение String-столбцов в одну группу:
cols = data.columns.tolist()
cols = cols[9:11] + cols[27:28] + cols[31:39] + cols[104:105] + cols[112:113] + cols[116:117] + cols[150:151] + cols[:9] + cols[11:27] + cols[28:31] + cols[39:104] + cols[105:112] + cols[113:116] + cols[117:150] + cols[151:]
data = data[cols]
data
```

```
# Кодирование столбцов с String-значениями
subdata = data.iloc[:, :15]

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
subdata_encoded = subdata.apply(le.fit_transform)
```

```
subdata_encoded
```

```
subdata_2 = data.iloc[:, 15:]
```

```
subdata_2
```

```
subdata_encoded.reset_index(drop=True, inplace=True)
subdata_2.reset_index(drop=True, inplace=True)
data_2 = pd.concat([subdata_encoded, subdata_2], axis = 1)
```

```
data_2
```

Нормировка данных так, чтобы среднее = 0, стандартное отклонение = 1

```
# Среднее значение
mymean = data_2.mean(axis=0)
# Стандартное отклонение
mystd = data_2.std(axis=0)
data_2 -= mymean
data_2 /= mystd
data_2
```

```
data_2 = data_2.drop(columns=['incineration_raion'])
data_2
```

```
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_2, ydata, test_size=0.33, random_state=42)
```

```
%tensorflow_version 2.x
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras import utils
from google.colab import files
from kerastuner.tuners import RandomSearch, Hyperband, BayesianOptimization
import numpy as np

# Модель 1 - 3 слоя, (128, 64, 32) нейронов в слоях соответственно
def build_model1(hp):
  model = Sequential()

  model.add(Dense(units=hp.Int('units_input', 
    min_value=128,
    max_value=128,
    step=32), input_shape=(X_train.shape[1],), activation='relu'))
  
  model.add(Dense(units=hp.Int('units_hidden', min_value=64, max_value=64, step=32), activation='relu'))
  model.add(Dense(units=hp.Int('units_hidden', min_value=32, max_value=32, step=32), activation='relu'))

  model.compile(optimizer="adam", loss='mse', metrics=['mae'])
  
  return model

# Модель 2 - 3 слоя, (32, 64, 128) нейронов в слоях соответственно
def build_model2(hp):
  model = Sequential()

  model.add(Dense(units=hp.Int('units_input', 
    min_value=32,
    max_value=32,
    step=32), input_shape=(X_train.shape[1],), activation='relu'))
  
  model.add(Dense(units=hp.Int('units_hidden', min_value=64, max_value=64, step=32), activation='relu'))
  model.add(Dense(units=hp.Int('units_hidden', min_value=128, max_value=128, step=32), activation='relu'))

  model.compile(optimizer="adam", loss='mse', metrics=['mae'])
  
  return model

# Модель 3 - 2 слоя, по 64 нейрона в слоях
def build_model3(hp):
  model = Sequential()

  model.add(Dense(units=hp.Int('units_input',
    min_value=64, 
    max_value=64,
    step=32), input_shape=(X_train.shape[1],), activation='relu'))
  
  model.add(Dense(units=hp.Int('units_hidden', min_value=64, max_value=64, step=32), activation='relu'))

  model.compile(optimizer="adam", loss='mse', metrics=['mae'])
  
  return model

tuner1 = BayesianOptimization(
  build_model1, # функция создания модели
  objective ='val_mae', # метрика, которую нужно оптимизировать
  max_trials = 5, # максимальное количество запусков обучения
  directory = 'test_directory1' # каталог, куда сохраняются обученные сети
)

tuner2 = BayesianOptimization(
  build_model2,
  objective ='val_mae',
  max_trials = 5,
  directory = 'test_directory2'
)

tuner3 = BayesianOptimization(
  build_model3,
  objective ='val_mae',
  max_trials = 5,
  directory = 'test_directory3'
)
```

```
tuner1.search(X_train, # Данные для обучения
             y_train, # Данные для обучения
             batch_size=32, # Размер мини-выборки
             epochs=10, # Количество эпох обучения
             validation_split=0.2, # Часть данных, которая будет использоваться для проверки
            verbose=1,)
```

```
tuner1.search_space_summary()
```

```
tuner1.results_summary()
```

```
models1 = tuner1.get_best_models(num_models=3)
```

```
for model in models1:
  model.summary()
  model.evaluate(X_test, y_test)
  print()
```
